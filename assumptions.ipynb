For SEM, need to check the normality and lineraty of the observed variables (Betsy's book)

# For China

1. check normality

Significant means not normal.

Shows like:
Statistics=0.870, p=0.000
R1: Sample does not look Gaussian (reject H0)

Need to drop na (but I do not want to now, because even for age, the most like normal variable is significant), or it would show like this:
Statistics=nan, p=1.000
S2: Sample looks Gaussian (fail to reject H0)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# assuming you have loaded your dataframe as ssea_duration
ssea = pd.read_csv('China/C_demo.csv')

for col in ssea.columns:
    # check normality
    stat, p = shapiro(ssea[col])
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    # interpret
    alpha = 0.05
    if p > alpha:
        print(col + ': Sample looks Gaussian (fail to reject H0)')
    else:
        print(col + ': Sample does not look Gaussian (reject H0)')
    
    # Plot distribution
    plt.figure(figsize=(8,4))
    sns.histplot(ssea[col], kde=True)
    plt.title('Distribution for ' + col)
    plt.show()
```

```python
# check for linearity.
sns.scatterplot(x='S1', y='R1', data=ssea)

# To add a fitted line
sns.regplot(x='S1', y='R1', data=ssea)

plt.title('Scatter plot with fitted line')
plt.show()

# Seems pretty flat
```

```python
# check outlier using boxplot

plt.figure(figsize=(20,10)) # You can adjust the figure size here
ssea.boxplot()
plt.xticks(rotation=90) # This will rotate x-axis labels for better visibility
plt.show()
```

```python
# check missing data

# count the number of missing values in each column
missing_values = ssea.isnull().sum()

# print the result
print(missing_values)

# Calculate the percentage of missing values in each column
missing_values_percentage = (ssea.isnull().sum() / len(ssea)) * 100

# Print the result
print(missing_values_percentage)
```
